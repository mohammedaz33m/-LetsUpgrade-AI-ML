## Logistic Regression: 1. Bank Loan Modelling

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("Bank_Personal_Loan_Modelling.xlsx",sheet_name=1)
data.head()
```

```python
data.info()
```

```python
data.describe().T
```

```python
data.columns
```

```python
y=data['Personal Loan'] #defining the dependent variable
```

```python
x=data[[ 'Age', 'Experience', 'Income', 'Family', 'CCAvg',
       'Education', 'Mortgage', 'Securities Account',
       'CD Account', 'Online', 'CreditCard']] # all the independent variables
```

```python
import statsmodels.api as sm 
```

```python
x1=sm.add_constant(x) #adding the b0 (beta not) constant value to the equation
```

```python
logistic = sm.Logit(y,x1)
```

```python
result=logistic.fit()
```

```python
result.summary()
```

__From the above table we can infer that__:

- those values with __p_value__ as __0__ are most significant in deciding the outcome of the dependent variable like, Const, Income, Family, Education, CD Account, online, Credit Card.


- other values which are __<0.05__ are significantly important & those __>0.05__ are insignificant (Age, Experience,Mortagage)

 

### Probability of getting the loan 

```python
# checking the probability with least significant IDV
# For Age lets check the loan outcome & assume the age =28
import math
prob = 1/(1+(math.exp(12.1928 +0.0536*28)))
prob
```

__The probability for someone with age 28 is 0.00011299 % which is very less__

```python
# checking the probability with most significant IDV
# For Age lets check the loan outcome & assume the income=155
import math
prob1 = 1/(1+(math.exp(12.1928 -0.0546*150)))
prob2 = 1/(1+(math.exp(12.1928 -0.0546*240)))
prob3 = 1/(1+(math.exp(12.1928 -0.0546*300)))
print(prob1)
print(prob2)
print(prob3)
```

__We can observe here that, as the income values increases from 150, 240 to 300 the probability of getting a loan also increases accordingly from 1.79% >> 71.3% >> to 98.5% respectively__
